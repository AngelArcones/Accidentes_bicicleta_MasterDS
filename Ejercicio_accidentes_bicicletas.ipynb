**Juntar todas las hojas en un solo excel**  

Primero, creo un DataFrame cargando la primera hoja  

Después, uso un *for loop* para ir cargando cada hoja una a una, a un objeto independiente,  
y lo combino con el DataFrame inicial  

Uso un *if* para cargar de forma diferente la del año 2013 del resto (por el numero de filas a omitir)

```python
full_excel = pd.read_excel("./Accidentes_sucio.xls",
                        sheet_name= by_sheet_names[0],
                        skiprows=7,
                        index_col=0,
                        usecols="A:K",
                        skipfooter=1)
full_excel["AÑO"] = by_sheet_names[0]

for i in range(1,len(by_sheet_names)):
    if by_sheet_names[i] == '2013' :
        by_year = pd.read_excel("./Accidentes_sucio.xls",
                        sheet_name= by_sheet_names[i],
                        skiprows=4,
                        index_col=0,
                        usecols="A:K",
                        skipfooter=1)
        by_year["AÑO"] = by_sheet_names[i]
        
    else:
        by_year = pd.read_excel("./Accidentes_sucio.xls",
                        sheet_name= by_sheet_names[i],
                        skiprows=7,
                        index_col=0,
                        usecols="A:K",
                        skipfooter=1)
        by_year["AÑO"] = by_sheet_names[i]
    full_excel = pd.concat([full_excel, by_year], axis=0, sort=False)
        
full_excel
```

```python
colnames = pd.Series(full_excel.columns)
colnames
```

Las columnas necesitan nombres mejores (tienen muchos espacios, tildes...)

Primera prueba sustituyendo

```python
colnames.str.replace('  ', '').str.replace(' ', '_')
```

Funciona, pero aún hay que eliminar algunos al final y quitar las tildes

Ahora, modificando el objeto (Series) con el listado de nombres

```python
colnames = colnames.str.replace('  ', '').str.replace(' ', '_').str.replace('Ó', 'O').str.replace('Í', 'I').str.replace('Ú', 'U')
colnames = colnames.apply(lambda x: x[:-1]+x[-1:].replace('_', ''))
colnames
```

Ahora si, lo meto como nuevos nombres

```python
full_excel.columns = colnames
full_excel.head(2)
```
